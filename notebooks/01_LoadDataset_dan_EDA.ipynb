## 7. Missing Value Heatmap

```python
import matplotlib.pyplot as plt
import seaborn as sns

check_datasets = {
    'wisman_gab'     : wisman_gab.iloc[:, :5],
    'wisnus (sample)': wisnus.iloc[:, :8] if 'wisnus' in dir() else pd.DataFrame(),
    'tpk_bintang'    : tpk_bintang if 'tpk_bintang' in dir() else pd.DataFrame(),
    'inflasi'        : inflasi if 'inflasi' in dir() else pd.DataFrame(),
    'usd_idr'        : usd_idr.head(50) if 'usd_idr' in dir() else pd.DataFrame(),
}
check_datasets = {k:v for k,v in check_datasets.items() if not v.empty}

n = len(check_datasets)
fig, axes = plt.subplots(1, n, figsize=(4*n, 4))
if n == 1: axes = [axes]

for ax, (name, df) in zip(axes, check_datasets.items()):
    sns.heatmap(df.isnull(), ax=ax, cbar=False, yticklabels=False,
                cmap=['#1e3a5f','#ef4444'])
    ax.set_title(name, fontsize=10, fontweight='bold')
    missing_pct = df.isnull().mean().mean() * 100
    ax.set_xlabel(f'Missing: {missing_pct:.1f}%', fontsize=8)

plt.suptitle('Missing Value Heatmap — Semua Dataset', fontsize=12, fontweight='bold', y=1.02)
plt.tight_layout()
plt.show()

# Summary tabel
print('=== Summary Missing Values ===')
for name, df in check_datasets.items():
    total_missing = df.isnull().sum().sum()
    total_cells   = df.shape[0] * df.shape[1]
    print(f'  {name:25s}: {total_missing:4d}/{total_cells} ({total_missing/total_cells*100:.1f}%)')
```

## 9. Stationarity Test (ADF)
Uji Augmented Dickey-Fuller untuk mengecek apakah data time-series stasioner.
Model LSTM/forecasting butuh data yang stasioner atau sudah di-differencing.

```python
try:
    from statsmodels.tsa.stattools import adfuller
except ImportError:
    import subprocess
    subprocess.run(['pip','install','statsmodels','-q'])
    from statsmodels.tsa.stattools import adfuller

def adf_test(series, name):
    series_clean = series.dropna()
    if len(series_clean) < 20:
        print(f'  {name}: ⚠️  data terlalu sedikit ({len(series_clean)} baris)')
        return
    result   = adfuller(series_clean, autolag='AIC')
    p_val    = result[1]
    adf_stat = result[0]
    is_stat  = p_val < 0.05
    status   = '✅ STASIONER (p<0.05)' if is_stat else '⚠️  NON-STASIONER (p≥0.05)'
    print(f'  {name:35s}: ADF={adf_stat:7.3f}  p={p_val:.4f}  → {status}')

print('=== ADF STATIONARITY TEST ===')
print()

# Buat series wisman dari wisman_gab
try:
    wg = wisman_gab.copy().iloc[:, :2]
    wg.columns = ['date', 'wisman']
    wg['wisman'] = pd.to_numeric(wg['wisman'], errors='coerce')
    adf_test(wg['wisman'], 'wisman (raw)')
    adf_test(wg['wisman'].pct_change(), 'wisman (growth MoM)')
    adf_test(wg['wisman'].diff(), 'wisman (1st difference)')
except Exception as e:
    print(f'  Error: {e}')

if 'usd_idr' in dir():
    try:
        usd_series = pd.to_numeric(
            usd_idr.iloc[:,1].astype(str).str.replace(',',''), errors='coerce').dropna()
        adf_test(usd_series, 'USD/IDR (raw)')
        adf_test(usd_series.pct_change(), 'USD/IDR (pct_change)')
    except Exception as e:
        print(f'  USD/IDR error: {e}')

print()
print('ℹ️  Non-stasioner → lakukan differencing atau pct_change sebelum LSTM/ARIMA')
print('ℹ️  Random Forest & Isolation Forest: OK dengan non-stasioner')
```
