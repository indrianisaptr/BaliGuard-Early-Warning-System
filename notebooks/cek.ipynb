```python
import pandas as pd
import os

files = [
    # isi dengan path semua file data dan notebook kamu
    '../data/final/master_dataset_clean.parquet',
    '../data/final/predictions_final.csv',
    '../data/final/crisis_dataset_final.csv',
    '../data/final/narratives_cache.json',
    '../data/final/predictions_final.csv'
    # dst...
]

for path in files:
    print(f"\n{'='*50}")
    print(f"FILE: {path} ({os.path.getsize(path)//1024} KB)")
    
    if path.endswith('.csv'):
        df = pd.read_csv(path, nrows=3)
    elif path.endswith('.parquet'):
        df = pd.read_parquet(path)
        df = df.head(3)
    
    print(f"Shape : {df.shape}")
    print(f"Kolom : {df.columns.tolist()}")
    print(f"Dtypes:\n{df.dtypes.to_string()}")
    print(f"Sample:\n{df.head(2).to_string()}")
```
